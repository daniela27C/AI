```python
# -*- coding: utf-8 -*-
"""Tiny Llama Example code from https://www.kaggle.com/code/mauriciofigueiredo/tiny-llama-example/notebook
with modifs Gabriel Turinici 22 oct 2024
"""
!pip install -q -U transformers
!pip install -q -U accelerate
!pip install -q -U bitsandbytes
import torch
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import pipeline

torch.set_default_device('cuda')
pipe = pipeline("text-generation", model="TinyLlama/TinyLlama-1.1B-Chat-v1.0", torch_dtype=torch.bfloat16, device_map="auto")
# We use the tokenizer's chat template to format each message - see https://huggingface.co/docs/transformers/main/en/chat_templating
messages = [
    {
        "role": "system",
        "content": "You are a friendly chatbot who always responds in the style of a pirate",
    },
    {"role": "user", "content": "How many helicopters can a human eat in one sitting?"},
]
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
print(outputs[0]["generated_text"])

messages = [
    {"role": "user", "content": "The sentiment of the sentence \"This code example is good enough to show how to use LLMs.\" is positive, neutral or negative?"},
]
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.1, top_k=50, top_p=0.95)
print(outputs[0]["generated_text"])

messages = [
    {"role": "user", "content": "O que ĂŠ um helicoptero? Explique com caracterĂ­sticas tĂŠcnincas prĂłprios desse tipo de transporte."},
]
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
print(outputs[0]["generated_text"])

# TODO : 1/ test other examples; 2/ test twice same 3/ make a chat engine to understand the context
```
